# Data visualisation task 2

```python
# Import libraries
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load data
wine = pd.read_csv('wine.csv')
```

### Histogram - Cabernet Sauvignon, Pinot Noir and Chardonnay

```python
# Histogram of Points per Wine Variety

cab_sauv = wine[wine['variety'] == 'Cabernet Sauvignon']
pinot_noir = wine[wine['variety'] == 'Pinot Noir']
chardonnay = wine[wine['variety'] == 'Chardonnay']

fig, ax = plt.subplots(3)

# Create histograms for each wine variety
ax[0].hist(cab_sauv['points'], bins = 20, color = 'red')
ax[0].set_title('Cabernet Sauvignon')
ax[0].set_xlabel('Points')
ax[0].set_ylabel('Frequency')

ax[1].hist(pinot_noir['points'], bins = 20, color = 'purple')
ax[1].set_title('Pinot Noir')
ax[1].set_xlabel('Points')
ax[1].set_ylabel('Frequency')

ax[2].hist(chardonnay['points'], bins = 20, color = 'yellow')
ax[2].set_title('Chardonnay')
ax[2].set_xlabel('Points')
ax[2].set_ylabel('Frequency')

# Add labels
leg_cab_sauv = mpatches.Patch(color = 'red', label = 'Cabernet Sauvignon')   # Create a cab sauv proxy artist for the legend
leg_pinot_noir = mpatches.Patch(color = 'purple', label = 'Pinot Noir')   # Create a pinot noir proxy artist for the legend
leg_chardonnay = mpatches.Patch(color = 'yellow', label = 'Chardonnay')   # Create a chardonnay proxy artist for the legend
plt.legend(handles = [leg_cab_sauv, leg_pinot_noir, leg_chardonnay])   # Pass the proxy artists to the legend

# Position the legend to the right of the plot area
plt.legend(handles=[leg_cab_sauv, leg_pinot_noir, leg_chardonnay], loc='upper left', bbox_to_anchor=(1.05, 1))

plt.tight_layout()  # Adjust the subplots to fit into the figure area
plt.show()
```
