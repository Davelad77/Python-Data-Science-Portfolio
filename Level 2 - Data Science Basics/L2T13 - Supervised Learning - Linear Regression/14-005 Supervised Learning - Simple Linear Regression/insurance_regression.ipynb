### Use the data in the relevant columns of insurance.csv to determine how age affects insurance costs:<br>
Plot a scatter plot with age on the x-axis and charges on the y-axis.<br>
Using linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on the data.<br>
Plot another scatter plot with the best-fit line.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Import the data
insurance_data = pd.read_csv("insurance.csv")
insurance_data.head()
```

```python
# The iloc function allows use to select rows from the dataframe for plotting.  
x = insurance_data.iloc[:,:1].values  # Select first column (age)
y = insurance_data.iloc[:,-1].values  # Select last column (charges)

plt.scatter(x,y,color = 'b')
plt.xlabel('Age')
plt.ylabel('Insurance Charge')
plt.title('Insurance Charge vs Age')
plt.show()
```

```python
# Fit the model
insurance_data = LinearRegression()
insurance_data.fit(x,y)
```

```python



# Plot the data and model
y_pred = insurance_data.predict(x)

plt.scatter(x,y,color = 'b')
plt.plot(x,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Insurance Charge')
plt.title('Insurance Charge vs Age with Model')
plt.show()
```
